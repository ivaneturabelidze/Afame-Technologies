```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read and show excel data
data = pd.read_excel('ECOMM DATA.xlsx')
data
```

```python
# Show information about dataframe's columns
data.info()
```

```python
# Filter data to get the most recent year's data
last_year_data = data[data['Order Date'] >= (data['Order Date'].max() - pd.DateOffset(years=1))]
last_year_data
```

```python
# Count total sales and profit of every year and the most recent year
total_sales = data['Sales'].sum()
total_profit = data['Profit'].sum()
last_year_sales = last_year_data['Sales'].sum()
last_year_profit = last_year_data['Profit'].sum()
print('Total Sales is $' + str('{:,}'.format(round(total_sales, 2))))
print('Total Profit is $' + str('{:,}'.format(round(total_profit, 2))))
print('Total Sales in the last year is $' + str('{:,}'.format(round(last_year_sales, 2))))
print('Total Profit in the last year is $' + str('{:,}'.format(round(last_year_profit, 2))))
```

```python
# Resample data to get sales in 6 month intervals
sales_data = data.resample('6M', on='Order Date')['Sales'].sum()
# Plot sales data using line chart
plt.plot(sales_data, marker='o')
plt.xticks(rotation=45)
plt.title('Sales over time by 6 month intervals')
plt.xlabel('Time')
plt.ylabel('million sales')
plt.grid(True)
plt.show()
```

```python
# Iterate through every year
for year in data['Order Date'].sort_values().dt.year.unique():
    # Get data for each year
    year_data = data[data['Order Date'].dt.year == year]
    # Get sales sums by categories
    sales_by_category = year_data.groupby('Category')['Sales'].sum()
    # Plot each year's popular categories with percentage
    plt.pie(sales_by_category, labels=sales_by_category.index, autopct='%1.1f')
    plt.title(f'Popular categories in {year}')
    plt.show()
```

```python
# Iterate through every year
for year in data['Order Date'].sort_values().dt.year.unique():
    # Get data for each year
    year_data = data[data['Order Date'].dt.year == year]
    # Get sales sums by categories
    sales_by_subcategory = year_data.groupby('Sub-Category')['Sales'].sum()
    # Contain categories which have less than 5% of sales in "other" category
    other_categories = sales_by_subcategory[sales_by_subcategory / sales_by_subcategory.sum() < 0.05]
    other_sales = sales_by_subcategory[other_categories.index].sum()
    sales_by_subcategory = sales_by_subcategory.drop(other_categories.index)
    sales_by_subcategory['other'] = other_sales
    # Plot popular categories of each year in pie charts
    plt.pie(sales_by_subcategory, labels=sales_by_subcategory.index, autopct='%1.1f')
    plt.title(f'Popular sub-categories in {year}')
    plt.show()
```
